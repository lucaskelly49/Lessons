# 1: Load Libraries

```python
# For dataframe manipulation

import pandas as pd

# For confusion matrix visualization

from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
import numpy as np
```

# 2: Look at an example dataset

```python
# Load and explore the data

df = pd.read_csv('UmpComp_Lesson.csv')
```

```python
# This data set shows features of a pitch in an MLB baseball game. The most important parts are the 
# True labels: des (this was the call made by the umpire)
# Predicted labels: comp_call (this is the predicted call by the computer)

# Splicing the neccessary columns here:

df = df[['ab', 'inning', 'pitchIdx', 'frame', 'batter', 'pitchtype', 'des', 'comp_call']]
```

```python
df
```

```python
# This data shows an entire game, but let's just look at Zack Greinke, the pitcher for the 
# Houston Astros.
# Greinke pitched through batter number 41, Aaron Judge, in the bottom of the 5th inning. 

df_Greinke = df.loc[(df['frame']=='bottom')&(df['ab']<=42)]
df_Greinke
```

# 3: Separate predictions from true labels

```python
# Separate predictions from true observations

y_true = df_Greinke['des'] 
y_pred = df_Greinke['comp_call']
```

# 4: Build Confusion Matrix

```python
# Function to visualize confusion matrix

def show_cf(y_true, y_pred, class_names=None, model_name=None):
    cf = confusion_matrix(y_true, y_pred)
    plt.imshow(cf, cmap=plt.cm.Blues)
    
    if model_name:
        plt.title("Confusion Matrix: {}".format(model_name))
    else:
        plt.title("Umpire vs. Computer - Confusion Matrix")
    plt.ylabel('Umpire')
    plt.xlabel('Computer')
    
    class_names = set(y_true)
    tick_marks = np.arange(len(class_names))
    if class_names:
        plt.xticks(tick_marks, class_names)
        plt.yticks(tick_marks, class_names)
    
    thresh = cf.max() / 2.
    
    for i, j in itertools.product(range(cf.shape[0]), range(cf.shape[1])):
        plt.text(j, i, cf[i, j], horizontalalignment='center', color='white' if cf[i, j] > thresh else 'black')

    plt.colorbar()
```

```python
show_cf(y_true, y_pred)
```

# 5: Evalute 

#### What is a positive?

Because we are looking at this from the view of the pitcher, let's consider a strike or 'called strike' as a positive. Let's consider the 'Ball' call being a negative. 

Don't get confused, we are considering the Umpire as the true label and the Computer as the predicted label. Though the computer may be making the most accurate call, the Umpire is the one who has the say in the game, therefore, should be considered as truth. 

#### True Positives: 14

This means that both the umpire and the computer made the same call, making it true. In this case, they both called a strike, making it positive. So, we have 14 true positives.

#### True Negatives: 33

This means that both the umpire and the computer made the same call, making it true. In this case, they both called a ball, making it negative. So, we have 33 true negatives.

#### False Positives: 1

This means that the umpire and the computer made different calls, making it false. In this case, the umpire called a ball when the computer called a strike. This is positive because the computer falsely called a positive, or a strike.

#### False Negatives: 5

This means that the umpire and the computer made different calls, making it false. In this case, the umpire called a strike when the computer called a ball. This is negative because the computer falsely called a negative, or a ball.
